## Wstępne czyszczenie danych

```python
to_drop = ["EmployeeCount", "StandardHours", "Over18"]

df.drop(columns=to_drop, inplace=True)
```

### sprawdzenie duplikatów

jezeli wszsytkie kolumny sie powtarzaja, uznaje to za duplikat - zostawiam pierwsze wystapienie danego rekordu

```python
duplicated_rows = df.duplicated(keep='first')
df[duplicated_rows]
```

```python
# usuniecie duplikatow
df = df[~duplicated_rows]
df
```

### sprawdzenie i ogarniecie outliers

```python
sns.boxplot(data=df, y="DistanceFromHome")
plt.show()
```

W przypadku kolumny DistanceFromHome, 75% wartosci jest rowna lub mniejsza 17. pozwala to przypuszczac ze uzyta jednostka są kilometry. Wartosci na poziomie 20_000 do 100_000 mozemy odrzucić, lub przujac zalozenie, ze wartosci dla czesci pracownikow zostaly wyrazone w innej jednostce, np metry.

opcja 1: zmiana jednostki (o ile to zalozenie jest prawidlowe),np  20_000[m] zmienimy na 20km, a 100_000[m] na 100km, co wydaje sie byc znacznie realniejszymi i wysoce prawodpodbnymi wartosciami. jednak wciaz pozostana wartosci znacznie odbiegajace od reszty na poziomie 1000km. Zakladajac brak pracy zdalnej, malo prawdopodobne by ktos dojezdzal tyle do pracy  
df.loc[df["DistanceFromHome"] > 1_000, "DistanceFromHome"] = df["DistanceFromHome"] / 1_000


opcja 2: przyjecie poziomu powyzej ktorego uznajemy wartosci za nieprawidlowe i usuwamy je z datasetu - przyjmuje 100km jako limit

```python
# opcja 2
df.drop(df[df["DistanceFromHome"] > 100].index, inplace=True)
```

```python
time_outliers = ["Age", "TotalWorkingYears", "YearsAtCompany", "YearsInCurrentRole", "YearsSinceLastPromotion", "YearsWithCurrManager"]

sns.boxplot(data=df[time_outliers])
plt.xticks(rotation=45)
plt.show()
```

```python
# usuniecie wartosci dla wieku, przepracowanych lat etc powyzej 100.
for col in time_outliers:
    df.drop(df.loc[df[col] > 100].index, inplace=True)


sns.boxplot(data=df[time_outliers])
plt.xticks(rotation=45)
plt.show()
```

```python
# usuniecie NaN 
df.dropna(inplace=True)
```

```python
# zmiana wartosci no / yes na 0 / 1
df["Attrition"].replace(["No", "Yes"], [0, 1], inplace=True)
```

```python
df.Attrition.value_counts()
```
